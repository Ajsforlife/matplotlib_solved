```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
```

```python
# Create a bar chart based upon the above data
fig, ax = plt.subplots()
ax.bar(cities,bars_in_cities,color="Blue")
```

```python
# Create the ticks for our bar chart's x axis
ax.set_xticks(np.arange(0, 5, step=1.0))  
```

```python
# Set the limits of the x axis
ax.set_xlim(-1,5)
```

```python
# Set the limits of the y axis
ax.set_ylim(0,9)
```

```python
# Give the chart a title, x label, and y label
ax.set_xlabel("Cities")
ax.set_ylabel("Bars per 10,000 households")
ax.set_title("Density of Bars per City")
```

```python
# Save an image of the chart and print it to the screen
fig, ax = plt.subplots()
ax.bar(cities,bars_in_cities,color="Blue")
ax.set_xticks(np.arange(0, 5, step=1.0))
ax.set_xlim(-1,5)
ax.set_ylim(0,9)
ax.set_xlabel("Cities")
ax.set_ylabel("Bars per 10,000 households")
ax.set_title("Density of Bars per City")
plt.savefig("Images/Bars_per_city.png")
plt.show()
```
